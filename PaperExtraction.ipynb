```python
artigo = "artigo.pdf"
```

```python
with pdfplumber.open(artigo) as pdf:
    texto_total = ""

    for pagina in pdf.pages:
        # Extrai blocos com coordenadas
        blocos = pagina.extract_words(
            keep_blank_chars=False,
            use_text_flow=True,
            extra_attrs=["x0", "top"]
        )

        # Ordena blocos pela posição horizontal (x0) e vertical (top)
        blocos_ordenados = sorted(blocos, key=lambda b: (b["top"], b["x0"]))

        # Concatena o texto ordenado em fluxo de leitura
        for b in blocos_ordenados:
            texto_total += b["text"] + " "

print(texto_total)
```
